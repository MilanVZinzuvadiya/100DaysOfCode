## Data Preprocess

```python
from sklearn import preprocessing
from pandas.api.types import is_string_dtype

le = preprocessing.LabelEncoder()

houseData = houseData.fillna('')

for i in houseData.columns:
    if is_string_dtype(houseData[i]):
        print(houseData[i],i)
        houseData[i] = le.fit_transform(houseData[i])
houseData.info()
```

## Train and Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(houseData.iloc[:,:-1],houseData['SalePrice'],test_size=0.25)
```

```python
x_train,x_test,y_train,y_test
```

## Model Creation and Training

```python
from sklearn.linear_model import LinearRegression
linearReg = LinearRegression()
linearReg.fit(x_train,y_train)

from sklearn.ensemble import RandomForestRegressor
RandomForest = RandomForestRegressor()
RandomForest.fit(x_train,y_train)
```
